```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('study_hours_result.csv')
data.head()
```

```python
x = data.iloc[:, :-1].values
y = data.iloc[:, 2].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.02, random_state=42)
```

```python
def distance(a, b):
    return np.sqrt(np.sum((a - b)**2))

def prediction(k, x_train, y_train, point):
    L = [distance(point, i) for i in x_train]
    indices = np.argsort(L)[: k]
    sum = 0
    for i in indices:
        sum += y_train[i]
    return sum/k
```

```python
prediction(9, x_train, y_train, x_test[0])
```

```python
y_test[0]
```

```python
total = 0
square_total = 0

for i in range(len(x_test)):
    pred = prediction(5, x_train, y_train, x_test[i])
    total += abs(pred - y_test[i])
    square_total += (pred - y_test[i])**2
    print(f"Predicted: {pred}, Actual: {y_test[i]}")

mae = total / len(x_test)
rmse = np.sqrt(square_total/len(x_test))
print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
def prediction_weight(k, x_train, y_train, point):
    L = [distance(point, i) for i in x_train]
    sorted_L = np.sort(L)[: k]
    indices = np.argsort(L)[: k]
    weights_sum = 0 
    sum = 0
    for i in sorted_L:
        weights_sum += 1 / i
    for i in range(k):
        sum += (1 / sorted_L[i]) * y_train[indices[i]]
    return sum / weights_sum
```

```python
prediction_weight(10, x_train, y_train, x_test[0])
```
